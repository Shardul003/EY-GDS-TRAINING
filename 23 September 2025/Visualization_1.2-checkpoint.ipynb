```python
import matplotlib.pyplot as plt
```

```python
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
#1. Basic Analysis
#Find the total revenue per product.
total_revenue=df.groupby("Product")["Revenue"].sum()
print(total_revenue)
```

```python
#Find the average units sold per product.
avg_unit=df.groupby("Product")["Units_Sold"].mean()
print(avg_unit)
```

```python
# 2. Visualization 1 – Line Chart
#Plot the Monthly Revenue Trend for each product (Jan–Mar).
obj=df.pivot(index="Month",columns="Product",values="Revenue")
obj.plot(kind="line",marker="o")
plt.title("Monthly Revenue Trend for each product (Jan–Mar)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
#3. Visualization 2 – Bar Chart
#Plot a bar chart of total revenue per product.
plt.bar(total_revenue.index, total_revenue.values)
plt.ticklabel_format(style="plain",axis="y")
plt.title("Total revenue per product.")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.grid(axis="y")
plt.show()
```

```python
#Visualization 3 – Pie Chart
#Plot a pie chart showing the contribution of each product to total revenue.

plt.pie(total_revenue.values,labels=total_revenue.index,autopct="%1.1f%%",startangle=140,colors=["orange","blue","green"])
plt.show()
```
